# 概率和分布

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import os
```

## 二项分布

```python
fig = plt.figure(figsize=(20,6))
fig.subplots_adjust(hspace=0.6,wspace=0.3)

for p in range(1,10):
    mybar = fig.add_subplot(3,3,p)
    mybar.bar(range(6),stats.binom.pmf(range(6),5,(p+0.0)/10),color="blue",edgecolor="black",alpha=0.5)
    mybar.set_xticks(range(6))
    mybar.set_xticklabels(range(6))
    plt.title("Figure4.1-p="+str((p+0.0)/10))
plt.show()
    
    
```

```python
stats.binom.pmf(range(6),5,0.5)
stats.binom.pmf(range(6),5,0.7)
```

## Poisson分布

```python
fig = plt.figure(figsize=(10,4))
plt.plot(range(25),stats.poisson.pmf(range(25),3),label=r'$\lambda=3$',linewidth=1,color='black',marker='o',markerfacecolor='black',markersize=6)
plt.plot(range(25),stats.poisson.pmf(range(25),6),label=r'$\lambda=6$',linewidth=1,color='green',marker='+',markerfacecolor='black',markersize=8,linestyle='--')
plt.plot(range(25),stats.poisson.pmf(range(25),10),label=r'$\lambda=10$',linewidth=1,color='orange',marker='+',markerfacecolor='black',markersize=10,linestyle=':')
plt.xlabel('k')
plt.ylabel('p(k)')
plt.title("Figure4.2-Poisson distribution")
plt.legend()
plt.show()
```

## 正态分布

```python
x = np.arange(-5,5,0.1)
fig = plt.figure(figsize=(20,6))
plt.plot(x,stats.norm.pdf(x,-2,0.5),linewidth=2,color='black',label="N(-2,0.5)")
plt.plot(x,stats.norm.pdf(x,0,1),linewidth=2,color='black',linestyle='--',label="N(0,1)")
plt.title("Figure4.3-norm distribution")
plt.legend()
plt.show()
```

## 均匀分布

```python
x = np.arange(-0.5,1.5,0.01)
fig = plt.figure(figsize=(20,6))
plt.plot(x,stats.uniform.pdf(x,0.1),linewidth=1,color='black',label=r"$F(3,20)$")
plt.title("Figure4.4-uniform distribution")
plt.legend()
plt.show()
```

# 本章节主要介绍seaborn第三方库的使用

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# 在Seaborn库中，hue参数用于按照特定的分组变量对数据进行可视化展示。
```

在Seaborn库中，hue参数用于按照特定的分组变量对数据进行可视化展示。当您在调用Seaborn函数时将hue参数设置为一个列名或变量名称时，该函数将根据该列或变量的唯一值创建不同的颜色分类，并将每个类别的数据点以不同的颜色呈现。这可以帮助您更好地理解数据之间的关系。

例如，在使用Seaborn绘制散点图时，您可能想要比较两个变量之间的关系，并且数据还有一个额外的分类变量（例如，性别）。通过将hue参数设置为性别列，您可以将散点图中男性和女性的数据点分别以不同的颜色表示，从而更好地了解性别在这两个变量之间的关系上扮演的角色。

```python
#data
tips = sns.load_dataset('tips')
tips.head()
```

```python
#distplot核密度分布图与曲线的绘制
sns.distplot(tips['total_bill'])
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#pairplot
sns.pairplot(tips)
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#boxplot
#sns.boxplot(x='day',y='total_bill',data=tips,palette='coolwarm')
sns.boxplot(x='day',y='total_bill',hue="smoker",data=tips,palette='coolwarm')
```

```python
#violinplot
sns.violinplot(x='day',y='total_bill',data=tips,palette='coolwarm')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
#heatmap
tips.corr()
#sns.heatmap(tips.corr())
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)
```

```python
titanic = pd.read_excel('titanic3.xls',index_col=False)
titanic
```

```python
#sns.jointplot(x='fare',y='age',data=titanic,kind='scatter')
sns.jointplot(x='fare',y='age',data=titanic,kind='reg')
```

```python
sns.distplot(titanic['fare'],kde=False,bins=30)
```

```python
sns.boxplot(x='pclass',y='age',data=titanic,palette='coolwarm')
```

```python
sns.swarmplot(x='pclass',y='age',data=titanic,palette='coolwarm')
```

```python
titanic.corr()
sns.heatmap(titanic.corr(),cmap='coolwarm',annot=True)
```

```python
g = sns.PairGrid(titanic)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(titanic)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
